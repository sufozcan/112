```python
import pandas as pd #importing libraries
import numpy as np
```

```python
dirty_data = pd.read_excel('Cleandata2.xlsx') # importing step 2 clean data
```

```python
dirty_data.head() # looking for imported true dataset and we saw that colab added again one unnecessary column and we dropped that column
```

```python
dirty_data.drop('Unnamed: 0', axis=1, inplace=True)
print(dirty_data)
```

```python
dirty_data.head() # Looking overall
```

```python
dirty_data.info() # looking for datatype and we saw it did not work so we change them last time
```

```python
dirty_data['BMI'] = dirty_data['BMI'].astype(float)
dirty_data['ID'] = dirty_data['ID'].astype(float)
dirty_data['Age'] = dirty_data['Age'].astype(float)
dirty_data['Glucose Level'] = dirty_data['Glucose Level'].astype(float)
dirty_data['Year'] = dirty_data['Year'].astype(float)
dirty_data['Exercise Duration'] = dirty_data['Exercise Duration'].astype(float)
```

```python
dirty_data.info()
```

```python
dirty_data.describe().transpose() # looking for NaN values filling but we noticed that we did not check outliers so we passed immediately to that part
```

```python
Q1 = dirty_data['Exercise Duration'].quantile(0.25) # Defining the quantiles
Q3 = dirty_data['Exercise Duration'].quantile(0.75)

IQR = Q3 - Q1 # Defining IQR

lower_limit = Q1 - 1.5 * IQR # Defining the lower and upper limits
upper_limit = Q3 + 1.5 * IQR

outliers = dirty_data[(dirty_data['Exercise Duration'] < lower_limit) | (dirty_data['Exercise Duration'] > upper_limit)] # defining and printing out outliers
outliers
```

```python
exercise_duration_mean = dirty_data['Exercise Duration'].mean() # We detected an outlier and we wanted to replaced with mean value so we printed it out
exercise_duration_mean
```

```python
dirty_data.loc[dirty_data['Exercise Duration'] == 180, 'Exercise Duration'] = 105 # In here, we wanted to deal with 105 except 106 beacuse of the other values are goes by multiplyig 3
```

```python
Q1 = dirty_data['Exercise Duration'].quantile(0.25) # After that part we looked one by one all of numerical variables which is having probability of outliers and we saw none of them have any outlier
Q3 = dirty_data['Exercise Duration'].quantile(0.75)

IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

outliers = dirty_data[(dirty_data['Exercise Duration'] < lower_limit) | (dirty_data['Exercise Duration'] > upper_limit)]
outliers
```

```python
Q1 = dirty_data['Glucose Level'].quantile(0.25)
Q3 = dirty_data['Glucose Level'].quantile(0.75)

IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

outliers = dirty_data[(dirty_data['Glucose Level'] < lower_limit) | (dirty_data['Glucose Level'] > upper_limit)]
outliers
```

```python
Q1_bmi = dirty_data['BMI'].quantile(0.25)
Q3_bmi = dirty_data['BMI'].quantile(0.75)


IQR_bmi = Q3_bmi - Q1_bmi


lower_limit_bmi = Q1_bmi - 1.5 * IQR_bmi
upper_limit_bmi = Q3_bmi + 1.5 * IQR_bmi


outliers_bmi = dirty_data[(dirty_data['BMI'] < lower_limit_bmi) | (dirty_data['BMI'] > upper_limit_bmi)]
outliers_bmi
```

```python
Q1_age = dirty_data['Age'].quantile(0.25)
Q3_age = dirty_data['Age'].quantile(0.75)


IQR_age = Q3_age - Q1_age


lower_limit_age = Q1_age - 1.5 * IQR_age
upper_limit_age = Q3_age + 1.5 * IQR_age


outliers_age = dirty_data[(dirty_data['Age'] < lower_limit_age) | (dirty_data['Age'] > upper_limit_age)]
outliers_age
```

```python
dirty_data.isnull().sum() # Now we wanted to passing through to filling NaN values by using isnull function to see any NaN values
```

```python
mode_val = dirty_data['Blood Pressure'].mode()[0] # Starting with categorical variable and filling it with mode
dirty_data['Blood Pressure'].fillna(mode_val, inplace=True)
```

```python
dirty_data.isnull().sum() # Checking the cleaning
```

```python
dirty_data.describe().transpose() # Looking descriptive stats for filling NaN values with them
```

```python
dirty_data['Age'].fillna(45, inplace=True) # Here we wanted to fill missing values one by one by looking at the descriptive stats table beacuse of it is so easy to see the values and filling by using that method
dirty_data['BMI'].fillna(26.7, inplace=True)
dirty_data['Glucose Level'].fillna(127, inplace=True)
dirty_data['Exercise Duration'].fillna(90, inplace=True)
```

```python
dirty_data.isnull().sum() # Checking missing values' cleaning step
```
