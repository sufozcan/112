```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Cleandata.xlsx')
```

```python
df.head()
```

```python
df.tail()
```

```python
sns.countplot(x='Year', hue='Heart Disease', data=df[df['Smoking Status'] == 'Smoker'])
plt.show()
```

```python
sns.violinplot(x='Physical Activity', y='BMI', data=df)
plt.xlabel('Physical Activity')
plt.ylabel('BMI')
plt.title('Relationship between BMI and Physical Activity')
plt.show()
```

```python
sns.catplot(x="Heart Disease",y="Age",data=df,kind="box",col="Gender")
```

```python
sns.catplot(x="Blood Pressure",y="BMI",data=df,kind="box",col="Smoking Status")
```

```python
plt.figure(figsize=(10, 6))
colors = {'Yes': 'green', 'No': 'red'}

sns.scatterplot(x='BMI', y='Glucose Level', data=df, hue='Heart Disease', palette=colors, s=100)
plt.title('Scatter Plot: BMI vs Glucose Level')
plt.xlabel('BMI')
plt.ylabel('Glucose Level')
plt.legend(title='Heart Disease', loc='upper right', labels=['No', 'Yes'])
plt.show()
```

```python
sns.catplot(x="Heart Disease",y="Age",data=df,kind="box",col="Gender")
plt.subplots_adjust(top=0.85)
plt.suptitle("Distribution of Ages by Gender and Heart Disease Status",fontsize=16)
plt.show()
```

```python
colors ={"Yes": 'red',"No":"green"}
sns.relplot(x="BMI",y="Glucose Level",data=df,kind="scatter",hue="Heart Disease",palette=colors)
plt.title(' BMI vs Glucose Level')
plt.show()
```

```python
sns.violinplot(x='Physical Activity', y='BMI', data=df)
plt.xlabel('Physical Activity')
plt.ylabel('BMI')
plt.title('Relationship between BMI and Physical Activity')
plt.show()
```

```python
heart_disease_yes = df[df['Heart Disease'] == 'Yes']
family_heart_count = heart_disease_yes.groupby(['Family History', 'Heart Disease', 'Gender']).size().reset_index(name='Count')
g = sns.catplot(x='Heart Disease', y='Count', data=family_heart_count, kind='bar', hue='Family History', col='Gender')
g.fig.subplots_adjust(top=0.85)
g.fig.suptitle('Family History and Heart Disease Counts by Gender', fontsize=16)
plt.show()
```

```python
bp_smoking_count = df.groupby(['Blood Pressure', 'Smoking Status']).size().unstack()
categories = ['Low', 'Normal', 'High']
colors = {"Non-smoker": "blue", "Smoker": "grey"}
bp_smoking_count = bp_smoking_count.reindex(categories)
bp_smoking_count.plot(kind='bar', stacked=True, rot=0, color=[colors[col] for col in bp_smoking_count.columns])
plt.title('Blood Pressure and Smoking Status Comparison')
plt.xlabel('Blood Pressure')
plt.ylabel('Count')
plt.legend(title='Smoking Status', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
